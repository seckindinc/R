```r
data("AirPassengers")
plot(AirPassengers)
#install.packages("forecast")
library(forecast)
```

```r
model1 <- auto.arima(AirPassengers, stepwise=FALSE, approximation=FALSE, parallel = TRUE, num.cores = NULL)
```

```r
model1
```

```r
summary(model1)
```

```r
forecast(model1,1)
```

```r
plot(forecast(model1,1))
```
